```python
# visualização de informação temporal com gráfico de linhas
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('USD_BRL_hist.csv', sep=',')
df['Data'] = pd.to_datetime(df['Data'], dayfirst=True)
df.set_index('Data', inplace=True)

mean_value = df['USD_BRL'].mean()  # cálculo da média

plt.figure(figsize=(15, 5))  # tamanho da figura

# Plota a série do dólar (plotar = traçar a linha na série temporal)
plt.plot(df.index, df['USD_BRL'], label='USD/BRL', color='black')  # linha do dólar

# Plota a linha da média do valor do dado central no gráfico
plt.axhline(mean_value, color='orange', linestyle='--', label=f'Média ({mean_value:.2f})')

# Cria máscaras para áreas acima e abaixo da média
above = df['USD_BRL'] > mean_value
below = df['USD_BRL'] <= mean_value

# Preencher áreas acima da média em vermelho claro
plt.fill_between(df.index, df['USD_BRL'], mean_value, where=above, 
                 color='red', alpha=0.3, label='Acima da Média')

# Preencher áreas abaixo da média em azul claro
plt.fill_between(df.index, df['USD_BRL'], mean_value, where=below, 
                 color='green', alpha=0.3, label='Abaixo da Média')

plt.title('USD/BRL - Valores e Períodos Acima/Abaixo da Média')
plt.xlabel('Data')
plt.ylabel('Valor (R$)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
