```python
# visualizao de informao temporal com grfico de linhas
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('USD_BRL_hist.csv', sep=',')
df['Data'] = pd.to_datetime(df['Data'], dayfirst=True)
df.set_index('Data', inplace=True)

mean_value = df['USD_BRL'].mean()  # clculo da mdia

plt.figure(figsize=(15, 5))  # tamanho da figura

# Plota a srie do dlar (plotar = traar a linha na srie temporal)
plt.plot(df.index, df['USD_BRL'], label='USD/BRL', color='black')  # linha do dlar

# Plota a linha da mdia do valor do dado central no grfico
plt.axhline(mean_value, color='orange', linestyle='--', label=f'Mdia ({mean_value:.2f})')

# Cria mscaras para reas acima e abaixo da mdia
above = df['USD_BRL'] > mean_value
below = df['USD_BRL'] <= mean_value

# Preencher reas acima da mdia em vermelho claro
plt.fill_between(df.index, df['USD_BRL'], mean_value, where=above, 
                 color='red', alpha=0.3, label='Acima da Mdia')

# Preencher reas abaixo da mdia em azul claro
plt.fill_between(df.index, df['USD_BRL'], mean_value, where=below, 
                 color='green', alpha=0.3, label='Abaixo da Mdia')

plt.title('USD/BRL - Valores e Perodos Acima/Abaixo da Mdia')
plt.xlabel('Data')
plt.ylabel('Valor (R$)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
