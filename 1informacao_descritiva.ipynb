```python
# visualização de informação com gráfico de barras de estatísticas descritivas
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('USD_BRL_hist.csv', sep=',')
df['Data'] = pd.to_datetime(df['Data'], dayfirst=True)  # convertendo a coluna 'Data' para o formato datetime

df.set_index('Data', inplace=True)  # definindo a coluna 'Data' como índice; necessário pro resample funcionar corretamente
df_resampled = df.resample('6ME').mean()  # calculando a média a cada 6 meses

x_labels = df_resampled.index.strftime('%m/%Y')  # formatando as datas para o eixo x
y_values = df_resampled['USD_BRL'].values  # valores reais (float)

plt.figure(figsize=(10, 3))  # definindo o tamanho da figura
plt.xticks(rotation=45)  # rotacionando os rótulos do eixo x para melhor legibilidade
plt.bar(x_labels, y_values, color='green', alpha=0.4)  # criando o gráfico de barras
plt.title('USD/BRL - Média a cada 6 meses')
plt.ylabel('Valor médio (R$)')
plt.ylim(bottom=0.5)  # definindo o limite inferior do eixo y
plt.grid(axis='y', linestyle='--', alpha=0.7)  # adicionando uma grade ao gráfico para facilitar a leitura
plt.show()
```
